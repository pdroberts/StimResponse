## Single unit analysis for DCN pharmacology protocol

The purpose of this analysis is to read and analyize the responses for single units for the following protocols: 
 * Response to a pure tone stimulus at the characteristic frequency and 10 dB below threshold.
 * Map the tuning curve including spectral-temporal histogram.
 * Map the response to broad-band noise to determine the threshold. 
 * Map the response to pure tones to determine the threshold.

Load data and map the spectral-temporal histograms for each attenuation level.

```python
unitNum = 47
test = 'Tuning'
stRaster, durs, tests = pf.GetTuningData(dataPath, mice, test, unitNum)
mouseName = mice.ix[unitNum]['Mouse ID'] + '_' + str(int(mice.ix[unitNum]['Depth']))
for c,t in enumerate(tests):
    ved.ShowSTH(stRaster[c], str(unitNum)+'_'+mouseName+'_'+str(t), duration=durs[c], figPath=[]);
```

Batch load the data and save plots of the spectral-temporal histograms.

```python
%%capture
LoadOrigData_PlotSTH = False  # Change to True to analyze data for first time
test = 'Tuning' # 'Washout Tuning' # 'Drug Tuning'  #'Washout Tuning'
tuningData = {}
if LoadOrigData_PlotSTH:
    for unitNum in mice.index[55:]:
        stRaster, durs, tests = pf.GetTuningData(mice, test, unitNum)
        mouseName = mice.ix[unitNum]['Mouse ID'] + '_' + str(int(mice.ix[unitNum]['Depth']))
        for c, t in enumerate(tests):
            if len(stRaster)>0:
                tuningData[str(unitNum)+'_'+mouseName+'_'+str(t)] = stRaster[c]
                print unitNum, mouseName, t
        cPickle.dump(tuningData, open(dirPath + 'tuningCurveData.p', 'wb'))
        for c,t in enumerate(tests):
            if len(durs) != 0 and durs[c] != 0:
                ved.ShowSTH(stRaster[c], str(unitNum)+'_'+mouseName+'_'+str(t), \
                            duration=durs[c], figPath='tuningCurves'+os.sep)            
else:
    tuningData.update(cPickle.load(open(dirPath + 'tuningCurveData.p', 'rb')));
```

Batch plotting of tuning curves.

```python
%%capture
SaveAllPlots = False    # Change to True to save plots and calculate tuning curve rates.
tuningCurves = {}
if SaveAllPlots:
    for t in tuningData.keys():
        testInfo = t.split('_')
        unitNum = int(testInfo[0])
        testNum = int(testInfo[-1])
        tuningCurves[t] = pf.PlotTuningCurves(mice, unitNum, \
                                              tuningData[t], testNum=testNum, \
                                              figPath=dirPath + 'tuningCurves'+os.sep)
        cPickle.dump(tuningCurves, open(dirPath + 'tuningCurves.p', 'wb'))
else:
    tuningCurves.update(cPickle.load(open(dirPath + 'tuningCurves.p', 'rb')));
```

### Broad band noise responses and threshold.

Load and plot the intensity-temporal histogram for IO tests. 

```python
tests = ['I/O BBN', 'Drug I/O BBN', 'Washout I/O BBN']
unitNumber = 47
autoRasters, dur =  pf.GetIOData(dataPath, mice, tests, unitNumber)
print autoRasters.keys()
for k in  autoRasters.keys():
    ved.ShowBBNTH(autoRasters[k], k, duration=dur, figPath=[]);
```

Batch analyze the IO tests.

```python
%%capture
LoadOrigData_IO_BBN = False    # Change to True to analyze data for first time
IORasters = {}
if LoadOrigData_IO_BBN:
    figDir = 'ioBBN'   # Directory in output path for figures 
    tests = ['I/O BBN', 'Drug I/O BBN', 'Washout I/O BBN']
    for n in range(0,len(mice)):
        mouse = mice.ix[n]
        if len(mouse['Mouse ID']) == 9: # and testNum != 0 and mouse['Type'] == 's':
            unitNumber = n
            autoRasters, dur = pf.GetIOData(dataPath, mice, tests, unitNumber)
            rasterSet = {}
            for k in  autoRasters.keys():
                rasterSet[k] = autoRasters[k]    
                ved.ShowBBNTH(autoRasters[k], k, duration=dur, figPath = os.sep+figDir+os.sep)
            IORasters[k.split('_')[0]+'_'+k.split('_')[1]+'_'+k.split('_')[2]] = rasterSet
            cPickle.dump(IORasters, open(dirPath + 'ioBBN.p', 'wb'))
else:
    IORasters.update(cPickle.load(open(dirPath + 'ioBBN.p', 'rb')));
```
