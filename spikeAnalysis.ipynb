## Single unit analysis for Batlab file type data 

The purpose of this analysis is to read and analyize the responses for a single unit for the following protocols: 
 * Response to a pure tone stimulus at the characteristic frequency and 10 dB below threshold.
 * Map the tuning curve including spectral-temporal histogram.
 * Map the response to broad-band noise to determine the threshold.
 * Map the response to two tones to determine responses to distortion products.
 * Map the responses to high frequency pure tones at 15 dB attenuation.
 * Responses to vocalizations presented at 15 dB attenuation.

```python
# 3/18/2015
%matplotlib inline
from IPython.display import display   # To print DataFrames nicely
import numpy as np
import scipy.stats as stats
import pandas as pd
import pymc as pm  # For Bayesian analysis
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize  # to choose figure size
import seaborn as sns                   # Custom graphing commands
import matplotlib.cm as cm                # Color map for STH (gray), etc
import matplotlib.gridspec as gridspec  # For layout of figures
import cPickle     # For saving and loading data structures from disk
import multiprocessing   # For efficient use of multicore hardware
from multiprocessing import Pool
dirPath = './data'    # location of Excel files
from StimResponse import *   # Custom data analysis commands
```

Load data on the units from the file containing data sheet information and initiate custom analysis objects.

```python
loadBatlab = True   # Change to True to re-analyze data or load first time
bra = BayesianResponseAnalysis.BayesianResponseAnalysis(dirPath)
ved = VocalEphysData.VocalEphysData(dirPath, loadBatlab)

# Load data sheet information
mice = ved.ReadMouseSheet( 'mice.xls' )
units = ved.GetUnits()
display(mice.tail())
```

Load and plot the response to a pure tone at CF and 10 dB below threshold.

```python
unitNum = 0
dep = units[unitNum].split('_')[1]
dep
thresh = mice[mice['depth']==int(dep)]['spk_threshold']
mice[mice['depth']==int(dep)]
```

Plot the recording electrode potential, detected spike raster and histogram.

```python
dataPath = './data/'
mouseStr = '1212' 
mouse_letter = 'a'
depth = '3174'
thresh = mice[mice['depth']==int(depth)]['spk_threshold']
thresh = 0.2
testName = 'histogram'
testNumber = mice[mice['depth']==int(depth)][testName].values[0]
dataFile = 'Mouse'+mouseStr+mouse_letter+'.pst'
spikeTrains, dur = ved.GetToneSpikeTrains(dataPath+dataFile, testNumber, spk_threshold=thresh, showNumTraces=10)

# display(spikeTrains)
#--- Plot raster ---
ved.PlotRaster(spikeTrains, title=units[unitNum], duration=dur*1000)
#--- Plot histogram ---
ved.PlotHisto(spikeTrains, title=units[unitNum], duration=dur*1000);
```

To perform a Bayesian estimation of the response, we import the a Markov chain Monte Carlo (MCMC) python module that implements the Metropolis-Hastings algorithm.

Define 4 parameters for the model of the response, $\lambda_1$ and $\lambda_2$ for two levels of activity, $\tau$ for the time when activity changes, and $\tau_2$ forthe duration of the deviation from spontaneous activity.

Construct the model of the response based on the three parameters, and run the Monto Carlo simulation on the data.

```python
figHisto = bra.BayesSpikeResponse(spikeTrains, duration=dur*1000, verbose=True )
```

**Frequency Tuning curve:**

Load data and map the spectral-temporal histograms for each attenuation level.

```python
testName = 'TuningCurve'
testNumber = mice[mice['depth']==int(dep)][testName].values[0]
spk_thresh=0.3
stRaster = ved.GetToneAutoTestData(dataPath+dataFile, testNumber, spk_threshold=0.3)
```

```python
test = 'TuningCurve'
mouseRange = range(unitNum,unitNum+1)
print 'Number of tuning curve tests:', len(stRaster)
ved.ShowSTH(stRaster, units[unitNum], duration=250);
```

Compute the frequency responses using Bayesian methods.

```python
stResponse, stResponseProb = bra.SpecTempResponseLoop(stRaster, duration=250, verbose=False, filePath=[])
```

Generate Fequency tuning curves and frequency response area maps for the units with tuning curve data.

Effect size: $d = \frac{\lambda_1 - \lambda_2}{\sigma_{ave}}$, where $\sigma_{ave} = \sqrt{(\sigma_1^2 + \sigma_2^2 )/2}$

```python
# measure in ['resProb', 'vocalResMag', 'vocalResMag_MLE', 'effectSize', 'effectSize_MLE', 'spontRate', 'spontRateSTD', 'responseLatency', 'responseLatencySTD', 'responseDuration']
measureNum = 4
for unit in [stResponseProb.keys()[0]]:
    bra.PlotFrequencyTuningCurves(stResponseProb, measureNum, unit=units[unitNum], filePath=[])
    bra.PlotFrequencyResponseArea(stResponseProb, measureNum, unit=units[unitNum], filePath=[]);
```

Plot the Bayesian estimation for each frequency for a unit and attenuation.

```python
attn = 0
label = units[unitNum] + ',  ' + stResponse.keys()[attn]
bra.PlotSTResponseEst(stResponse[stResponse.keys()[attn]], label, firstFreq=1);
```
